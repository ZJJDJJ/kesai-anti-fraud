```python

from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
import pandas as pd
df_transform = pd.read_csv('data/train_transform.csv', index_col= 'PERSONID')
```

```python
df_train = df_transform.loc[df_transform.LABEL != 2,:]
df_test = df_transform.loc[df_transform.LABEL ==2, :]
df_test.drop(['LABEL'], axis = 1, inplace= True)

Y = df_train.LABEL
X = df_train.drop(['LABEL'], axis = 1)
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
print(X.shape)
X_train.head()
```

```python

# é€»è¾‘å›žå½’æ¨¡åž‹
log_model = LogisticRegression()
# è®­ç»ƒé€»è¾‘å›žå½’æ¨¡åž‹
log_model.fit(X_train, y_train)
 
# é¢„æµ‹yçš„å€¼
y_pred = log_model.predict(X_test)
 
# æŸ¥çœ‹æµ‹è¯•ç»“æžœ
print(metrics.confusion_matrix(y_test, y_pred))
print(metrics.classification_report(y_test, y_pred))
```

```python
log_model_submit = LogisticRegression()
# è®­ç»ƒé€»è¾‘å›žå½’æ¨¡åž‹
log_model_submit.fit(X, Y)
 
# é¢„æµ‹yçš„å€¼
y_pred = log_model_submit.predict(df_test)
```
